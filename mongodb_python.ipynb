```python
student_id_filter = input("Enter Student ID (or leave blank for all): ")
course_id_filter = input("Enter Course ID (or leave blank for all): ")

match_filter = {}
if student_id_filter.strip():
    match_filter["student_id"] = int(student_id_filter)
if course_id_filter.strip():
    match_filter["course_id"] = int(course_id_filter)

pipeline = []
if match_filter:
    pipeline.append({"$match": match_filter})

# Add the same $lookup/$unwind/$project stages as before
pipeline += [
    {"$lookup": {
        "from": "students",
        "localField": "student_id",
        "foreignField": "student_id",
        "as": "student_info"
    }},
    {"$unwind": "$student_info"},
    {"$lookup": {
        "from": "courses",
        "localField": "course_id",
        "foreignField": "course_id",
        "as": "course_info"
    }},
    {"$unwind": "$course_info"},
    {"$project": {
        "_id": 0,
        "student_name": "$student_info.name",
        "course_name": "$course_info.course_name",
        "instructor": "$course_info.instructor"
    }}
]

results = list(df_e.aggregate(pipeline))
for r in results:
    print(f"{r['student_name']} → {r['course_name']} ({r['instructor']})")

```

```python
from pymongo import MongoClient
import pandas as pd

enrollments = pd.read_csv("./enrollments.csv")
df_e.delete_many({})  # optional: clear previous enrollments

for rec in enrollments.to_dict(orient="records"):
    df_e.insert_one(rec)

client = MongoClient("mongodb://localhost:27017/")
db = client["school_db"]

pipeline = [
    # Lookup student info
    {"$lookup": {
        "from": "students",
        "localField": "student_id",
        "foreignField": "student_id",
        "as": "student_info"
    }},
    {"$unwind": "$student_info"},

    # Lookup course info
    {"$lookup": {
        "from": "courses",
        "localField": "course_id",
        "foreignField": "course_id",
        "as": "course_info"
    }},
    {"$unwind": "$course_info"},

    # Group by student and collect courses in a list
    {"$group": {
        "_id": "$student_info.name",
        "courses": {"$push": "$course_info.course_name"}
    }},

    # Project final output
    {"$project": {
        "_id": 0,
        "student_name": "$_id",
        "courses": 1
    }}
]

results = list(db.enrollments.aggregate(pipeline))

# Print nicely
for r in results:
    print(f"{r['student_name']} → {', '.join(r['courses'])}")
```

```python
#11. Write a Python function to search for a student by name (partial matches allowed using LIKE).
from pymongo import MongoClient
client = MongoClient("mongodb://localhost:27017/")
db = client["school"]
partial_name = input("Enter student name (partial search): ")
query = {"name": {"$regex": partial_name, "$options": "i"}}
students = db.students.find(query, {"_id": 0})
found = False
print("\nSearch results:\n")
for student in students:
    found = True
    print("ID:", student.get("student_id"),
          "| Name:", student.get("name"),
          "| Age:", student.get("age"),
          "| Grade:", student.get("grade"),
          "| Email:", student.get("email"))
if not found:
    print("No matching student found.")
```

```python
#18.Handle exceptions in Python to gracefully catch duplicate entries (e.g., inserting a student with the same ID)
from pymongo import MongoClient, errors

client = MongoClient("mongodb://localhost:27017/")
db = client["school"]
db.students.create_index("student_id", unique=True)


records = [{"student_id": i, "name": f"Student_{i}"} for i in range(1, 101)]

for rec in records:
    try:
        db.students.insert_one(rec)
        print(f"Inserted: {rec['student_id']} - {rec['name']}")
    except errors.DuplicateKeyError:
        print(f"Duplicate found! Student ID {rec['student_id']} already exists.")
```

```python

#19. Write a program to implement transactions in Python:
# Insert a student
# Enroll them in a course
# Roll back if the second query fails
with client.start_session() as session:
    try:
        with session.start_transaction():
            student = {"student_id": "10", "name": "leo", "age": 45, "grade": "C"}
            db.students.insert_one(student, session=session)
            enrollment = {"enroll_id": "6000", "student_id": "10", "course_id": "8999"}
            db.enrollments.insert_one(enrollment, session=session)
            print("Transaction successful! Student and enrollment inserted.")
    except errors.PyMongoError as e:
        print("Transaction aborted due to error:", e)
```
