```python
#2. Create a function in Python to create the tables (students, courses, enrollments) in school_db.
def create_tables():
    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password="tamrapathi3383",
        database="schools_db"
    )
    cursor = conn.cursor()

    cursor.execute("""
    CREATE TABLE IF NOT EXISTS students (
        student_id INT PRIMARY KEY,
        name VARCHAR(100),
        age INT,
        grade VARCHAR(5),
        email VARCHAR(100)
    )""")

    cursor.execute("""
    CREATE TABLE IF NOT EXISTS courses (
        course_id INT PRIMARY KEY,
        course_name VARCHAR(100),
        instructor VARCHAR(100)
    )""")

    cursor.execute("""
    CREATE TABLE IF NOT EXISTS enrollments (
        enroll_id INT PRIMARY KEY,
        student_id INT,
        course_id INT,
        FOREIGN KEY (student_id) REFERENCES students(student_id),
        FOREIGN KEY (course_id) REFERENCES courses(course_id)
    )""")
    print("âœ… Tables created")

create_tables()
```

```python
#17.Create a stored procedure in MySQL (e.g., GetStudentCourses(student_id)) and call it from Python.
import mysql.connector

# Establish connection to the database
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="tamrapathi3383",
    database="schools_db"  # Make sure database is selected
)

mycursor = conn.cursor()

# Create the stored procedure using Python
proc = """
CREATE PROCEDURE GetStudentCourses(IN sid INT)
BEGIN
    SELECT s.name, c.course_name
    FROM enrollments e
    JOIN students s ON e.student_id = s.student_id
    JOIN courses c ON e.course_id = c.course_id
    WHERE s.student_id = sid;
END
"""

try:
    mycursor.execute("DROP PROCEDURE IF EXISTS GetStudentCourses")
    mycursor.execute(proc)
    conn.commit()
    print("Stored procedure 'GetStudentCourses' created successfully.")
except Exception as e:
    print("Error creating procedure:", e)

# Call the stored procedure
try:
    mycursor.callproc('GetStudentCourses', [1])  # Pass student_id = 1
    
    for result in mycursor.fetchall():
        print(result)
        
except Exception as e:
    print("Error calling procedure:", e)

finally:
    mycursor.close()
    conn.close()
```

```python
#18. Handle exceptions in Python to gracefully catch duplicate entries (e.g. inserting a student with the same ID). 
import mysql.connector
from mysql.connector import Error

def insert_student(student_id, name, age, grade):
    conn = None
    cursor = None
    try:
        conn = mysql.connector.connect(
            host="localhost", 
            user="root", 
            password="tamrapathi3383",  # Use correct password
            database="schools_db"  # Use correct database name
        )
        cursor = conn.cursor()
        cursor.execute(
            "INSERT INTO students (student_id, name, age, grade) VALUES (%s, %s, %s, %s)", 
            (student_id, name, age, grade)
        )
        conn.commit()
        print("âœ… Student inserted successfully")
        
    except mysql.connector.errors.IntegrityError as e:
        print(f"Duplicate entry found: student_id {student_id} already exists")
        
    except Error as e:
        print(f"Error: {e}")
        
    finally:
        if cursor is not None:
            cursor.close()
        if conn is not None and conn.is_connected():
            conn.close()

insert_student(1, "Surya", 21, "A")  # Example
```

```python
#19. Write a program to implement transactions in Python: 
# Insert a student 
# Enroll them in a course 
import mysql.connector

def transaction_example():
    conn = None
    cursor = None
    try:
        conn = mysql.connector.connect(
            host="localhost", 
            user="root", 
            password="tamrapathi3383",
            database="schools_db"
        )
        cursor = conn.cursor()
        
        # Start transaction
        conn.start_transaction()

        # Insert new student with explicit student_id
        cursor.execute("INSERT INTO students (student_id, name, age, grade) VALUES (%s, %s, %s, %s)", (888, "NewStudent", 20, "B"))

        # Enroll them in a course (use valid course_id)
        cursor.execute("INSERT INTO enrollments (enroll_id, student_id, course_id) VALUES (%s, %s, %s)", (888, 888, 1))

        conn.commit()
        print("Transaction successful")
    except Exception as e:
        if conn is not None:
            conn.rollback()
        print("Transaction failed Rolled back.", e)
    finally:
        if cursor is not None:
            cursor.close()
        if conn is not None and conn.is_connected():
            conn.close()

transaction_example()
```

```python
#20.Close the database connection safely in Python, ensuring both cursor and connection are properly closed.
import mysql.connector
from mysql.connector import Error

def safe_close():
    conn = None
    cursor = None
    try:
        conn = mysql.connector.connect(
            host="localhost", 
            user="root", 
            password="tamrapathi3383",  # Use correct password
            database="schools_db"  # Use correct database name
        )
        cursor = conn.cursor()
        cursor.execute("SELECT COUNT(*) FROM students")
        count = cursor.fetchone()[0]
        print(f"âœ… Students count: {count}")
        
    except Error as e:
        print(f"âŒ Error: {e}")
        
    finally:
        # Safely close cursor
        if cursor is not None:
            cursor.close()
            print("âœ… Cursor closed")
        
        # Safely close connection
        if conn is not None and conn.is_connected():
            conn.close()
            print("âœ… Connection closed safely")

safe_close()
```
